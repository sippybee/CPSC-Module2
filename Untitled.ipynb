```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
alist = ["banana", "orange","apple","banana", "orange"]
```

```python
le = LabelEncoder()
```

```python
le.fit_transform(alist)
```

```python
df = pd.DataFrame()
df["Data"] = alist
df["LE"] = le.fit_transform(alist)
```

```python
gd = pd.get_dummies(df["Data"],dtype=int)
```

```python
pd.concat([gd,df],axis=1)
```
