```python
df["IW_fl"] = df["Index Weight"].apply(lambda x : float(x.strip("%")))
```

```python
av_weigt = round(df["IW_fl"].mean(),2)
av_weigt
f"Average weight of a stock in the index is {av_weigt}%"
```

```python
# What is the total Volume?
df["Volume2"] = df["Volume"].apply(lambda x : float(x.strip(" M")))
df["Volume2"].sum()
```

```python
df["Vol_fl"] = df["Volume"].str.strip("M").astype(float)
```

```python
df
# how many stock have Last price less than 100 
df[df["Last"] < 100].count().iloc[0]
```
