```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame()
df["Area"] = [2600, 3000, 3200, 3600, 4000]
df["Price"] = [550000, 565000, 610000, 680000, 725000]
```

```python
# Let's predict price 3300 and 5000
```

```python
plt.scatter(df["Area"], df["Price"], c="red")
plt.ylabel("Price")
plt.xlabel("Area")
# price = m * area + b
```

```python
X = df[["Area"]].values
X.ndim
```

```python
y = df["Price"]
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[3300]])
```

```python
m = model.coef_[0]
```

```python
b = model.intercept_
```

```python
y = m*3300 + b
y
```

```python
predicted = model.predict(X)
```

```python
plt.scatter(df["Area"], df["Price"], c="red")
plt.scatter(3300, model.predict([[3300]]), marker="*",c="red", s=400)
plt.vlines(3300, ymin=525000, ymax=model.predict([[3300]]), color="red", linestyles=":")
plt.hlines(model.predict([[3300]]), xmin=2500,xmax=3300, color="red", linestyles=":")
plt.annotate("y = m*3300 + b",(3350, 630715))
plt.plot(df["Area"], predicted)
plt.ylabel("Price")
plt.xlabel("Area")
```
