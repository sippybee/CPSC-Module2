```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier,GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
```

```python
y = iris.target
```

```python
iris.keys()
```

```python
tree_model_l = []
forest_model_l = []
AdaBoos_model_l = []

for i in range(50):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2)
    tree_model = DecisionTreeClassifier()
    tree_model.fit(X_train,y_train)
    tree_model_l.append(tree_model.score(X_test,y_test))
    forest_model = RandomForestClassifier()
    forest_model.fit(X_train,y_train)
    forest_model_l.append(forest_model.score(X_test,y_test))
    AdaBoos_model = AdaBoostClassifier()
    AdaBoos_model.fit(X_train,y_train)
    AdaBoos_model_l.append(AdaBoos_model.score(X_test,y_test))
```

```python
r = pd.DataFrame()
r["tree_model"] = tree_model_l
r["forest_model"] = forest_model_l
r["AdaBoos_model"] = AdaBoos_model_l
```

```python
r.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=43)
models = [
    ("KNN", KNeighborsClassifier()),
    ("DecisionTree", DecisionTreeClassifier()),
    ("GradientBoosting", GradientBoostingClassifier()),
    ("RandomForest", RandomForestClassifier()),
    ("NeuralNetwork", MLPClassifier(max_iter=2000))
]
 
for model_name, model in models:
    model.fit(X_train, y_train)
    accuracy = model.score(X_test, y_test)
    y_pred = model.predict_proba(X_test)
    roc_auc = roc_auc_score(y_test, y_pred[:, 1],multi_class='ovr')
    print(f"{model_name}: ROC = {roc_auc*100:.1f}%, Accuracy = {accuracy*100:.1f}%")
```
