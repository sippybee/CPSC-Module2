```python
df = pd.read_csv("Social_Network_Ads.csv")
```

```python
df["gender_code"] = df["Gender"].apply(lambda x: 1 if x=="Female" else 0)
```

```python
df.columns
```

```python
X = df[["Age", 'EstimatedSalary']]
y = df["Purchased"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
scaler_min_max = MinMaxScaler()
X_min_max_scaled = scaler_min_max.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_min_max_scaled, y, test_size=0.2, random_state=43)

models = (("KNN",KNeighborsClassifier()),
          ("LogisticReg",LogisticRegression()))

for model_name, model in models:
    model.fit(X_train,y_train)
    print(model_name, model.score(X_test,y_test))
```
