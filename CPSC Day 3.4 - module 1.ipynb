```python
import pandas as pd
```

```python
df = pd.read_csv("chipotle.csv")
```

```python
df.head()
# get rid of the col order_id
df.drop("order_id", axis = 1, inplace = True)
```

```python
df.columns
```

```python
df.rename(columns={"item_price":"order_price"}, inplace=True)
```

```python
# what is the most expensive order what was ordered?
df["order_price"] = df["order_price"].apply(lambda x : float(x.strip("$")))
```

```python
df[df["order_price"] == df["order_price"].max()]
```

```python
df['item_name'].nunique()
```

```python
# How many unique items are in the data set ---> item_name
df['item_name'].unique()
# how many times people order 'Chicken Bowl'
```

```python
df[df["item_name"]=='Chicken Bowl']["quantity"].sum()
```

```python
# # How many time people ordered more than 1 'Canned Soda'
# 1. get all orders of Canned Soda
cs = df[df["item_name"]=='Canned Soda']
# 2. Then check quantity
cs[cs["quantity"]>1].count().iloc[0]
```

```python
condition1 = df["item_name"].str.contains("canned soda", case=False)
condition2 = df["quantity"]>1
df[condition1 & condition2].count().iloc[0]
```

```python
# what is 'Chicken Bowl'total revenue
chicken_bowl_revenue=df[df["item_name"] == "Chicken Bowl"]["order_price"].sum()
```

```python
total_revenue = df["order_price"].sum()
```

```python
chicken_bowl_revenue_percent = chicken_bowl_revenue / total_revenue *100
chicken_bowl_revenue_percent
```

```python
revenue = df.groupby("item_name")[["order_price"]].sum()
revenue.sort_values("order_price", ascending=False, inplace=True)
```

```python
revenue["Per"] = revenue["order_price"]/total_revenue*100
revenue["Per"] = revenue["Per"].round(2)
revenue
```
